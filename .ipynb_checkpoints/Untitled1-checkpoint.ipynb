```python
%matplotlib inline
import pygame
import gym
import numpy as np
import matplotlib.pyplot as plt
import random
from IPython import display
from tensorflow import keras
import math
import time
```

```python
import tensorflow as tf
```

```python
class Agent:
    def __init__(self, obs_shape, act_size):
        self.obs_shape = obs_shape
        self.act_size = act_size
    def network (self, train = True, node_size = 64, a_function = tf.keras.layers.LeakyReLU()):
        inputs = tf.keras.Input(shape=(self.obs_shape,),name="input")
        x = tf.keras.layers.Dense(node_size, activation=a_function,name="dense_1")(inputs)
        x = tf.keras.layers.Dense(node_size, activation=a_function,name="dense_2")(x)
        outputs = tf.keras.layers.Dense(self.act_size, name="output")(x)
        model = tf.keras.models.Model(inputs = inputs, outputs=outputs, name="nn", trainable=train)
        return model

    def network_1(self,train=True):
        x = tf.keras.layers.Dense(64, activation=tf.keras.layers.LeakyReLU(),name="dense_1")(inputs)
        outputs = tf.keras.layers.Dense(self.act_size, name="output")(x)
        model = tf.keras.models.Model(inputs = inputs, outputs=outputs, name="nn", trainable=train)
        return model
    def network_4(self,train=True):
        inputs = tf.keras.Input(shape=(self.obs_shape,),name="input")
        x = tf.keras.layers.Dense(64, activation=tf.keras.layers.LeakyReLU(),name="dense_1")(inputs)
        x = tf.keras.layers.Dense(64, activation=tf.keras.layers.LeakyReLU(),name="dense_2")(x)
        x = tf.keras.layers.Dense(64, activation=tf.keras.layers.LeakyReLU(),name="dense_3")(x)
        x = tf.keras.layers.Dense(64, activation=tf.keras.layers.LeakyReLU(),name="dense_4")(x)
        outputs = tf.keras.layers.Dense(self.act_size, name="output")(x)
        model = tf.keras.models.Model(inputs = inputs, outputs=outputs, name="nn", trainable=train)
        return model
```

```python
class Util:
    def __init__(self, _decaysteps = 50, _decayrate = 0.9, max_history = 10000, batch_size = 200, tau = 0.08):
        self.optimizer = tf.keras.optimizers.Adam(learning_rate = tf.keras.optimizers.schedules.ExponentialDecay(0.01, decay_steps = _decaysteps, decay_rate = _decayrate))
        self.max_history = max_history
        self.replay_history = []
        self.current_experience = 0
        self.batch_size = batch_size
        self.tau = tau
        
    def save_experience(self, current_state, action, reward, next_state):
        #print(self.current_experience)
        if self.current_experience < self.max_history:            
            self.replay_history.append([current_state, action, reward, next_state])
        else:
            #print("AYIE -> ", self.current_experience % self.max_history)
            self.replay_history[self.current_experience % self.max_history] = [current_state, action, reward, next_state]
        self.current_experience += 1

    def sample_from_experience(self):
        if self.current_experience >= self.batch_size:
            batch = random.sample(self.replay_history, self.batch_size)
        else:
            batch = random.sample(self.replay_history, len(self.replay_history))
        return batch
    
    def td_loss(self, nn, discount = 0.99):
        loss = []
        sampled_experiences = self.sample_from_experience()       
        
        for current_state, action, reward, next_state in sampled_experiences:
            b_action = [0.0] * nn.output.shape[1]
            b_action[action] = 1.0
            b_action = tf.constant([b_action])
            #DOUBLE DQN
            #SELECTION OF THE ACTION DONE BY THE ONLINE WEIGHTS
            #ESTIMATE GREEDY VALUES WITH CURRENT 
            
            
            #get current Q value from network
            current_q = nn(tf.convert_to_tensor([current_state]))
            #find max q value for next state
            max_q = tf.math.reduce_max(nn(tf.convert_to_tensor([next_state])))
            loss.append(tf.math.square((reward + discount * max_q - current_q) * b_action))
        return tf.math.reduce_mean(loss, axis=0)
        
    def update_model(self, nn):
        #back prop
        with tf.GradientTape() as tape:
            loss = self.td_loss(nn)
            grads = tape.gradient(loss, nn.trainable_variables)
            self.optimizer.apply_gradients(zip(grads, nn.trainable_variables))
            return loss
```

```python

#activation function
total_rewards_af = []
total_losses_af = []
agents = [
    Agent(4,2).network(),
    Agent(4,2).network(a_function = tf.keras.activations.tanh),
    Agent(4,2).network(a_function = tf.keras.activations.softmax)
]
l = ["LeakyRelu","tanh","softmax"]
colors = ['g','b','r']
number_of_episodes = 2000
for idx, agent in enumerate(agents):
    env = gym.make('CartPole-v1')
    state = env.reset()
    #agent = Agent(4,2).network()
    utility = Util(batch_size = 128)
    start = time.time()
    n_e = number_of_episodes
    rewards = []
    e = 1
    min_e = 0.01
    i, early_stop = 0,0
    tau = 20
    total_iterations = 0
    losses = []
    while i < number_of_episodes:
        current_state = env.reset()
        step = 0
        while True:
            total_iterations+=1
            r = np.random.uniform()
            #EXPLORATION VS EXPLOITATION
            if r < e:
                action = env.action_space.sample()
            else:
                #sample action from the network by picking highest output from the nn, convert to numpy
                action = tf.math.argmax(tf.reshape(agent(tf.convert_to_tensor([current_state])), [-1])).numpy()
            next_state, reward, done, info = env.step(action)
            step+=1
            utility.save_experience(current_state, action, reward, next_state)
            current_state = next_state

            if (total_iterations % tau) == 0:
                losses.append(np.mean(utility.update_model(agent)))

            #if (i % tau) == 0 :
                #utility.update_model(agent)
            #e decay
            e = max(e * 0.99, min_e)
            #env.render()

            if done:
                rewards.append(step)
                
                i+=1
                if step >= 500:
                    early_stop+=1
                else:
                    early_stop = 0
                if early_stop >= 10:
                    n_e = i
                    i = number_of_episodes                    
                break
    end = time.time()
    print(f"BATCH SIZE -> {l[idx]} finished in {n_e} episodes total time needed for training {end-start} (s)")
    plt.subplot(1,2,1)    
    plt.plot(np.arange(len(rewards)),rewards,c=colors[idx])
    plt.title("Rewards")
    plt.ylabel("Rewards")
    plt.xlabel("Episodes")
    plt.legend(["LeakyReLU","tanh","softmax"])
    plt.subplot(1,2,2)
    plt.plot(np.arange(len(losses)), losses, c=colors[idx])
    plt.title("Average Loss")
    plt.ylabel("Loss")
    plt.xlabel("Episodes")
    plt.legend(["LeakyReLU","tanh","softmax"])
    total_rewards_af.append(rewards)
    total_losses_af.append(losses)
plt.suptitle("Activation Function (LeakyReLU, tanh, softmax")
plt.show()
plt.savefig('figures/dqnactivationfunction', bbox_inches='tight')
```
